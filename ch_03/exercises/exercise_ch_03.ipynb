2 - With faang, use type conversion to cast the values of the date column into datetimes and the volume column into integers. Then, sort by date and ticker

```python
df = pd.read_csv('faang.csv')
```

```python
df.head()
```

```python
# convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])
```

```python
# convert 'volume' column to int format
df['volume'] = df['volume'].astype(int)
```

```python
# sort the values by date and ticker
df = df.sort_values(by=['date', 'ticker'], ascending=True)
df.head(10)
```

3 - Find the seven rows in faang with the lowest value for volume

6. Read in the covid19_cases

```python
# read the csv file
covid_df = pd.read_csv('covid19_cases.csv')
covid_df.head()
```

```python
# creat a new column for datetime
# the format in dataframe is dd / mm / yyyy
covid_df['date'] = pd.to_datetime(covid_df['dateRep'], format='%d/%m/%Y')
covid_df.head()
```

```python
covid_df.tail()
```

```python
# set the index column
covid_df.set_index('date', inplace=True)
```

```python
covid_df.head()
```

```python
# sort the index
covid_df.sort_index(ascending=True, inplace=True)
```

```python
covid_df.head()
```

```python
covid_df['countriesAndTerritories'] = covid_df['countriesAndTerritories'].replace(['United_States_of_America', 'United_Kingdom'], 
                                                                                  ['USA', 'UK'])
```

```python
covid_df[covid_df['countriesAndTerritories'] == "USA"].head()
```

```python
# filter the cleaned COVID-19 cases data down to selected countries
countries = ['Argentina', 'Brazil', 'China', 'Colombia', 
             'India', 'Italy', 'Mexico', 'Peru', 'Russia', 
             'Spain', 'Turkey', 'UK', 'USA']

covid_countries = covid_df[covid_df['countriesAndTerritories']
                           .isin(countries)]
```

```python
covid_countries.head()
```

```python
# pivot the data
covid_countries.head()
```

```python
covid_pivot = pd.pivot_table(covid_countries, 
                             index='date',
                             columns='countriesAndTerritories',
                             values='cases',
                             aggfunc='sum',  # Aggregate function to combine multiple entries for the same date/country
                             fill_value=0)   # Fills missing values with 0
```

```python
covid_pivot
```
